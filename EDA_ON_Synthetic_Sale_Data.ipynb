# Data Analysis

# Data Quality check

```python
print("\n==DATA QUALITY ASSESSMENT==\n")
#Check for missing values
print("\nMissing Values:\n")
missing_data=df.isnull().sum()
print(missing_data[missing_data>0])

#Check for duplicates
print(f"\nDuplicate Rows:{df.duplicated().sum()}")
```

# Data type Conversion

```python
# Converting date columns to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')

# Check conversion
print("\nDate columns converted:\n")

print(f"Order Date: {df['Order Date'].dtype}")
print(f"Ship Date: {df['Ship Date'].dtype}")
```

# Univariate analysis

```python
# Univariate Analysis is the samplest from of statistical analysis. As the name suggests--"uni"
#means "one," and "variate" means "variable"--it is the analysis of a single variable.
print("\n== CATEGORICAL VARIABLES ANALYSIS ==\n")
categorical_cols=['Customer Segment','Category','Region','Payment Mode','Delivery Status','Auto Reorder']
for col in categorical_cols:
    print(f"\n-- {col} --")
    print(df[col].value_counts())
    
    #Visualization
    plt.figure(figsize=(6,6))
    df[col].value_counts().plot(kind='bar',color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=30)
    plt.tight_layout()  #👉 Automatically adjusts the graph/plot elements properly.
    plt.show()
```

# Numerical Variables Analysis

```python
print("\n== NUMERICAL VARIABLES ===")
numerical_cols=['Quantity','Unit Price','Discount (%)','Sales Amount','Cost Price','Profit','Stock Left','Reorder Quantity']

for col in numerical_cols:
    print(f"\n-- {col} --\n")
    print(f"Mean: {df[col].mean():.2f}")
    print(f"Median: {df[col].median():.2f}")
    print(f"Standard Deviation: {df[col].std():.2f}")
    print(f"Minimum: {df[col].min():.2f}")
    print(f"Maximum: {df[col].max():.2f}")
    
    #Visualization
    plt.figure(figsize=(6,4))
    plt.subplot(1,2,1) #👉 1 row, 2 columns, 1st plot
    df[col].hist(bins=30,color='lightgreen',alpha=0.7)
    plt.title(f'Distribution of {col}')
    
    plt.subplot(1,2,2)  #👉 1 row, 2 columns, 2nd plot
    df.boxplot(column=col)
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()
    plt.show()
    
```

# Bivariate analysis

```python
#Bivariate Analysis is the statistical analysis of two variables at the same time.The core purpose is to investigate the relationship between them.
print("\n== SALES & PROFIT BY CATEGORY ==\n")
category_performance=df.groupby('Category').agg({'Sales Amount':'sum','Profit':'sum','Order ID':'count'}).rename(columns={'Order ID':'Order Count'})
category_performance['Profit Margin']=(category_performance['Profit']/category_performance['Sales Amount'])*100
print(category_performance,'\n\n')

#Visualization
fig,axes=plt.subplots(2,2,figsize=(15,12))

#Sales by Category
category_performance['Sales Amount'].plot(kind='bar',ax=axes[0,0],color='blue',alpha=0.7)
axes[0,0].set_title('Total Sales by Category') 
axes[0,0].tick_params(axis='x',rotation=45)

#**alpha = 0.7 =**
#👉 “The plot will be 70% visible and 30% transparent.”
#**tick_params =**#👉 “Change the style of the x-axis and y-axis ticks (axis numbers).”


#Profit by Category
category_performance['Profit'].plot(kind='bar',ax=axes[0,1],color='green',alpha=0.7)
axes[0,1].set_title('Total Profit by Category')
axes[0,1].tick_params(axis='x',rotation=45)


#Profit Margin
category_performance['Profit Margin'].plot(kind='bar',ax=axes[1,0],color='orange',alpha=0.7)
axes[1,0].set_title('Profit Margin by Category')
axes[1,0].tick_params(axis='x',rotation=45)


#Order Count
category_performance['Order Count'].plot(kind='bar',ax=axes[1,1],color='red',alpha=0.7)
axes[1,1].set_title('Order Count by Category')
axes[1,1].tick_params(axis='x',rotation=45)

plt.tight_layout()
plt.show()
```

# Regional Performance


```python
print("\n== SALES & PROFIT BY REGION ==\n")
regional_performance=df.groupby('Region').agg({'Sales Amount':'sum','Profit':'sum','Order ID':'count'}).rename(columns={'Order ID':'Order Count'})
regional_performance['Profit Margin']=(regional_performance['Profit']/regional_performance['Sales Amount'])*100
print(regional_performance,'\n\n')

#Visualization
plt.figure(figsize=(10,6))

#Sales by Region
plt.subplot(2,2,1)
regional_performance['Sales Amount'].plot(kind='pie',color='blue',autopct='%1.1f%%')
plt.title('Sales Distribution by Region')

#Profit by Region
plt.subplot(2,2,2)
regional_performance['Profit'].plot(kind='pie',color='green',autopct='%1.1f%%')
plt.title('Profit Distribution by Region')

#Order Count by Region
plt.subplot(2,2,3)
sns.barplot(x=regional_performance.index,y='Order Count',data=regional_performance)
plt.title('Order Count by Region')
plt.xticks(rotation=45)

#Profit Margin by Region
plt.subplot(2,2,4)
sns.barplot(x=regional_performance.index,y='Profit Margin',data=regional_performance)
plt.title('Profit Margin by Region')

plt.tight_layout()
plt.show()
```

# Time Series

```python
print("\n== TIME SERICES ANALYSIS ==\n")

#Extract time components
df['Order Month']=df['Order Date'].dt.to_period('M')
df['Order Year']=df['Order Date'].dt.year

#Monthly trends
monthly_trends=df.groupby('Order Month').agg({'Sales Amount':'sum','Profit':'sum','Order ID':'count'}).rename(columns={'Order ID':'Order Count'})
print("Monthly Trends:")
print(monthly_trends.tail(10),'\n')

#Visualization
plt.figure(figsize=(15,10))

plt.subplot(3,1,1)
monthly_trends['Sales Amount'].plot(kind='line',marker='*',color='blue')
plt.title('Monthly Sales Trends')
plt.ylabel('Sales Amount')
plt.grid(True)  #👉 plt.grid(True) means:“Show grid lines on the graph"

plt.subplot(3,1,2)
monthly_trends['Profit'].plot(kind='line',marker='*',color='green')
plt.title('Monthly Profit Trends')
plt.ylabel('Profit')
plt.grid(True)

plt.subplot(3,1,3)
monthly_trends['Order Count'].plot(kind='line',marker='*',color='orange')
plt.title('Monthly Order Count Trends')
plt.ylabel('Order Count')
plt.grid(True)
```

# Advanced analysis

```python
print("\n== CORRELATION ANALYSIS ==\n")

#Select numarical columns for correlation
numerical_df=df[['Quantity','Unit Price','Discount (%)','Sales Amount','Cost Price','Profit','Stock Left','Reorder Quantity']]
#Correlation matrix
correlation_matrix=numerical_df.corr()
print(correlation_matrix,'\n\n')

#Heatmap visualization
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',linewidths=0.5,square=True)
plt.title('Correlation Matrix Heatmap')
plt.tight_layout()
plt.show()
```

# Customer Segment Analysis

```python
print("\n== CUSTOMER SEGMENT ANALYSIS ==\n")

# Check exact column names
print(df.columns.tolist())

# Groupby Analysis
segment_analysis = df.groupby('Customer Segment').agg({
    'Sales Amount': ['sum', 'mean'],
    'Profit': ['sum', 'mean'],
    'Order ID': 'count'
}).round(2)

# Rename Columns
segment_analysis.columns = [
    'Total Sales',
    'Avg Sales',
    'Total Profit',
    'Avg Profit',
    'Order Count'
]

# Display Result
print(segment_analysis, '\n\n')


# ================= Visualization =================

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Total Sales
segment_analysis['Total Sales'].plot(
    kind='bar',
    ax=axes[0,0],
    color='blue',
    alpha=0.7
)
axes[0,0].set_title('Total Sales by Customer Segment')
axes[0,0].tick_params(axis='x', rotation=30)

# Total Profit
segment_analysis['Total Profit'].plot(
    kind='bar',
    ax=axes[0,1],
    color='green',
    alpha=0.7
)
axes[0,1].set_title('Total Profit by Customer Segment')
axes[0,1].tick_params(axis='x', rotation=30)

# Average Sales
segment_analysis['Avg Sales'].plot(
    kind='bar',
    ax=axes[1,0],
    color='orange',
    alpha=0.7
)
axes[1,0].set_title('Average Sales by Customer Segment')
axes[1,0].tick_params(axis='x', rotation=30)

# Average Profit
segment_analysis['Avg Profit'].plot(
    kind='bar',
    ax=axes[1,1],
    color='red',
    alpha=0.7
)
axes[1,1].set_title('Average Profit by Customer Segment')
axes[1,1].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```

# Key Insights ans summary

```python
print("\n== KEY INSIGHTS SUMMARY ==\n")
#Overall Metrics
total_sales=df['Sales Amount'].sum()
total_profit=df['Profit'].sum()
toral_orders=df['Order ID'].nunique()
avg_profit_margin=(total_profit/total_sales)*100


print(f"Overall performance:\n")
print(f"Total Sales: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Total Orders: {toral_orders}")
print(f"Average Profit Margin: {avg_profit_margin:.2f}%")


#Top Performing Categories
top_categories=df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
print(f"\nTop Performaing Categories:\n)")
for i, (category, profit) in enumerate(top_categories.items(), start=1):
    print(f"{i}. {category}: ${profit:,.2f}")


#Regional Performance
best_region=df.groupby('Region')['Profit'].sum().idxmax()
print(f"\nBest Performing Region: {best_region}")


#Delivery Performance
delivery_stats=df['Delivery Status'].value_counts(normalize=True)*100  #normalize=👉 It is used when we want to see the **percentage / ratio / proportion** instead of the **raw count**.
print(f"\nDelivery Status:")
for status, percentage in delivery_stats.items():
    print(f"{status}: {percentage:.2f}%")


#Customer Segment
best_segment=df.groupby('Customer Segment')['Profit'].sum().idxmax()
print(f"\nBest Performing Customer Segment: {best_segment}")    
```
